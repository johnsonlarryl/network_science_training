```python
import matplotlib.pyplot as plt
import networkx as nx
from numpy import random as nprand
import random
```

```python
# Sets the random seeds used by NetworkX, so that all diagrams will appear the same
# no matter how many times the cell is executed.
seed = hash("Network Science in Python") % 2**32
nprand.seed(seed)
random.seed(seed)
```

```python
# Sets the size of each the visualization figure 
plt.rcParams.update({'figure.figsize' : (7.5, 7.5)})
```

```python
G = nx.Graph()
```

```python
G.add_node('A')
```

```python
G.add_nodes_from(['B', 'C'])
```

```python
G.add_edge('A', "B")
```

```python
G.add_edges_from([('B', 'C'), ('A', 'C')])
```

```python
nx.draw_networkx(G)
```

```python
# Adds nodes and edges so that the network matches the example network
# Eliminates the need to call add_node() directly
G.add_edges_from([('B', 'C'), ('C', 'E')])
nx.draw_networkx(G)
```
